 # 판다스의 기초 
 

 
- 판다스는 무엇인가?

판다스(pandas)는 파이썬에서 사용하는 데이터분석 라이브러리이다. 

행과 열로 이루어진 데이터 객체를 만들어 다룰 수 있게 되며, 대용량의 데이터들을 처리하는데 매우 편리하다.

---

**예제1** pandas를 pd라는 이름으로, matplotlib.pyplot를 plt라는 이름으로 불러와서 사용한다. 

# 판다스 자료구조


데이터 프레임(DataFrame)은 데이터(data), 인덱스(index), 칼럼(columns)으로 구성된다. 



---

데이터프레임 생성 방법은 다음과 같다.

  |   활용법  |   설명   |
  |:--------:|:--------:|
  |   pandas.DataFrame(data, index, columns)    |   데이터, 인덱스, 칼럼으로 구성된 데이터프레임을 형성한다. |
  
---

딕셔너리의 활용법은 다음과 같다.  

  |  특징 | 표현방법  |   예  |
  |:--------:|:--------:| :--------:|
  |  자료를 key와 values에 대응하여 정리한다.   |  { }를 사용, 쉼표로 각 원소를 구분 | {‘a’ : ‘apple’, ‘b’ : ‘banana’} |

그래프를 그리는 방법은 다음과 같다. 

| 활용법 | 설명 | 
| :----: | :---:|
|plt.bar(데이터프레임['열이름'], 데이터프레임['열이름']) | 세로형 막대그래프를 그린다. | 
|plt.barh(데이터프레임['열이름'], 데이터프레임['열이름']) | 가로형 막대그래프를 그린다. | 
|plt.line(데이터프레임['열이름'], 데이터프레임['열이름']) | 꺽은선그래프를 그린다. |
